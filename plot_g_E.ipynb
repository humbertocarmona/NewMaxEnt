```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
data = np.genfromtxt("log_g.csv", delimiter=",", names=True)
E = data['E']
log_g = data['log_g']

# Sort by energy
sort_idx = np.argsort(E)
E_sorted = E[sort_idx]
log_g_sorted = log_g[sort_idx] - np.max(log_g)  # shift

# Plot
plt.figure(figsize=(8, 5))
plt.plot(E_sorted, log_g_sorted, marker='o')
plt.xlabel("Energy E")
plt.ylabel(r"$\log g(E) - \log g(E_\mathrm{max})$")
plt.title("Log Density of States")
plt.grid(True)
plt.tight_layout()
plt.show()
```
